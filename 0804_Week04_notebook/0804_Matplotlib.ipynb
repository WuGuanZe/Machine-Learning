## 基本函數畫法

```python
x = np.linspace(-10,10,100)   # x 範圍, 就 -10 至 10, 給 100 個點
```

```python
y = np.sin(x)   # y 是 sin(𝑥)
```

```python
plt.plot(x,y)
```

## 函數圖

```python
y = np.sin(5*x) / (1+x**2)
```

```python
plt.plot(x,y)
```

## 改變顏色

```python
x = np.linspace(-10,10,200)
```

```python
y = np.sinc(x)
```

```python
plt.plot(x,y,'r')
```

## 改變線條

```python
x = np.linspace(-10,10,200)
```

```python
y = np.cos(x)
```

```python
plt.plot(x,y,'r-.')
```

```python
plt.plot(x,y, c='r', lw=3)
```

## 參數式圖形

```python
r = 3
```

```python
t = np.linspace(-2*np.pi, 2*np.pi, 200)
```

```python
x = r*np.cos(t)
y = r*np.sin(t)
```

```python
plt.plot(x, y, lw=3)
```

```python
ax = plt.gca()
ax.set_aspect('equal')
plt.plot(x, y, lw=3)
```

## 特別的曲線

## subplot 一次畫多個圖

```python
x = np.linspace(0, 10, 100)
```

```python
plt.subplot(221)
plt.plot(x, np.sin(x), c='#e63946', lw=3)

plt.subplot(222)
plt.plot(x, np.sin(3*x), c='#7fb069', lw=3)

plt.subplot(223)
plt.scatter(x, np.random.randn(100), c='#daa73e', s=50, alpha=0.5)

plt.subplot(224)
plt.bar(range(10), np.random.randint(1,30,10), fc='#e55934')
```

## 圖表進階色彩表示法-練習

```python
x = np.linspace(-10,10,200)
```

```python
y = np.sinc(x)
```

```python
plt.plot(x,y, c = '#00a676', lw=3)  # lw 表示線的粗細寬度
```

## Mark 表示突顯原來的點

```python
x = range(20)
```

```python
y = np.random.randn(20)
```

```python
plt.plot(x,y) 
```

```python
# 若加上 marker='0'
plt.plot(x, y, marker='o')
```

## Mark 範例-練習1

```python
plt.plot(x, y, c='#6b8fb4', lw=5, marker='o', mfc='#fffa7c', mec="#084c61", mew=3, ms=20)
```

## Mark 範例-練習2

```python
x = np.linspace(0, 10, 200)
```

```python
y = np.sin(5*x) / (1+x**2)
```

```python
#此未加 markevery
plt.plot(x, y, lw=3, marker='o', ms=10)
```

```python
# 在此加 markevery
plt.plot(x, y, lw=3, marker='o', ms=10, markevery=10)
```

## 雙色長條圖

```python
x = np.arange(1,6)
```

```python
plt.bar(x - 0.4, [3, 10, 8, 12, 6], width=0.4, ec='none',
fc='#e63946')
plt.bar(x, [6, 3, 12, 5, 8], width=0.4, ec='none', fc='#7fb069')
```

回到我們的例子, 我們來算總級分, 注意加的方向!

排序是這樣做, 再次注意這不會改變原來的 DataFrame。

```python
df.sort_values(by="區域別", ascending=False).head()
```

```python
df2 = df.sort_values(by=["區域別", "人口密度"], ascending=False)
```

```python
df2.head()
```

重設 index!

```python
df2[df2.區域別 == '高雄市鳳山區']
```

也可以這多重條件! 注意括號是必要的, 而 `&` 是 and 運算, `|` 是 or 運算。

```python
df2[df2.區域別 > '高雄']
```

```python
df2[(df2.區域別>'高雄市') & (df2.人口密度 < 10000)]
```

```python
df2[(df2.區域別 < '高雄市') | (df2.人口密度 < 10000)]
```

```python
df2.drop("年底人口數總計", axis=1, inplace=True)
```

```python
df2.head()
```

列的話是要用那一列的 index。

```python
df2.drop(5)
```

這樣其實有點不實際, 因為我們還要自己找出要刪資料的 index! 
下面的方式雖然麻煩, 但應該比較實用。

```python
df2.drop(df2[df2.區域別 == "高雄市鳳山區"].index)
```

```python
df2[df2.區域別 != "高雄市鳳山區"]
```
